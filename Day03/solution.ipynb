Problem-01:
Write a Python program that accepts a comma-separated sequence of words as input and prints the words in a comma-separated sequence after sorting them alphabetically. Here's an example:

 Input: "dog,cat,apple,banana"
Output: "apple,banana,cat,dog"

```python
#with split and sorted
```

```python
string = "dog,cat,apple,banana"
splitedText = string.split(',') # ['dog', 'cat', 'apple', 'banana']
```

```python
sortedList = sorted(splitedText) # ['apple', 'banana', 'cat', 'dog']
```

```python
sortedString = ','.join(sortedList)
```

```python
print(sortedString)
```

A palindrome is a word, phrase, number, or other sequence of characters that reads the same forward and backward. Here's an example:
Input: "racecar"
Output: True


```python
giveWord = 'book'
reversWord = giveWord[::-1]
if giveWord == reversWord:
    print(True)
else:
    print(False)
```

Problem-04:
This problem requires you to write a Python program that calculates the factorial of a given number. 
The factorial of a non-negative integer n, denoted by n!, is the product of all positive integers less than or equal to n. For example, 5! = 5 * 4 * 3 * 2 * 1 = 120. Here's an example:
Input: 5
Output: 120
Explanation: 5! = 5 * 4 * 3 * 2 * 1 = 120.


```python
# range method
```

```python
givenNum = 7
```

```python
factorial = 1
for n in range(1,givenNum+1):
    product = n * factorial
    factorial = product

print(factorial)
```

An anagram is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once. Here's an example:

Input: "listen", "silent"
Output: True
Explanation: "listen" and "silent" are anagrams of each other

```python
# by sorted method
# by forloop 
```

```python
str1 = "listen"
str2 = "silent"
```

Problem-07:
Write a python program to check if a number is prime number or not.
A prime number is a positive integer greater than 1 that has no positive integer divisors other than 1 and itself. For example, 2, 3, 5, 7, 11, and 13 are prime numbers. On the other hand, 4, 6, 8, 9, and 12 are not prime numbers. Here's an example:

Input: 5
Output: True
Explanation: 5 is a prime number because it is only divisible by 1 and itself.

```python
# with for loop
```

```python
givenNum = 5
```

```python
if givenNum > 2:
    for x in range(2,givenNum+1):
        if int(givenNum/x) == 0:
            print(False)
            break
        
    else:
        print(True)
else:
    print(False)
```
